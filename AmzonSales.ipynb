**Importing Required Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Importing the dataset**


```python
sales_data= pd.read_csv('Amazon Sales data.csv')
sales_data
```

```python
# checking the shape
sales_data.shape
```

```python
# Show columns present in dataframe
sales_data.columns
```

```python
# Show basic information of dataframe
sales_data.info()
```

```python
# Changing the data type of different column
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])
sales_data['Ship Date'] = pd.to_datetime(sales_data['Ship Date'])
```

```python
sales_data['Region'] = sales_data['Region'].astype(str)
sales_data['Country'] = sales_data['Country'].astype(str)
sales_data['Item Type'] = sales_data['Item Type'].astype(str)
sales_data['Sales Channel'] = sales_data['Sales Channel'].astype(str)
sales_data['Order Priority'] = sales_data['Order Priority'].astype(str)
```

```python
# Checking no. of Null Values
sales_data.isnull().sum()
```

```python
sales_data.describe()
```

```python
# Adding extra column to dataframe which contain only month, year and month with year
sales_data['Order Month'] = sales_data['Order Date'].dt.month
sales_data['Order Year'] = sales_data['Order Date'].dt.year
sales_data['Order Date MonthYear'] = sales_data['Order Date'].dt.strftime('%Y-%m')
```

```python
sales_data.head()
```

```python
#save this to csv file
sales_data.to_csv('Amazon Sales transformed data.csv', index=False)
print('data saved to csv')
```

```python
df=sales_data
```

**Exploratory Data Analysis**

```python
#groupby region,sales channel, total profit
df.groupby(['Region','Sales Channel'])['Total Profit'].sum()
```

```python
#year wise sales
Year_sales= df.groupby('Order Year')['Total Revenue'].mean()
plt.figure(figsize=(8,3))
sns.barplot(x=Year_sales.index,y=Year_sales.values)
plt.title('Average sales by Year')
plt.xlabel('Year')
plt.ylabel('total revenue')
plt.xticks(rotation=90)
plt.show()
```

```python
#pie chart of total profit in region wise
plt.figure(figsize=(6,6))
region_TotalRevenue= df.groupby('Region')['Total Revenue'].mean()
plt.pie(region_TotalRevenue,labels=region_TotalRevenue.index,autopct='%1.1f%%')
plt.title('Average profit in region wise')
plt.show()
```

```python
#groupby Total revenue by item types
TotalRevenue_ItemTypes=df.groupby('Item Type')['Total Revenue'].sum()
```

```python
#bar chart for Total Revenue by Item types
plt.figure(figsize=(6,4))
TotalRevenue_ItemTypes.plot(kind='bar')
plt.title('Total Revenue by Item Types')
plt.xlabel('Item Types')
plt.ylabel('Total Revenue')
plt.show()
```

```python
#group total revenue by sales channel
TotalRevenue_SalesChannel=df.groupby('Sales Channel')['Total Revenue'].mean()
```

```python
#pie chart for total revenue by sales channel
plt.figure(figsize=(6,6))
plt.pie(TotalRevenue_SalesChannel,labels=TotalRevenue_SalesChannel.index,autopct='%1.1f%%')
plt.title('Total Revenue by Sales Channel')
plt.show()
```

```python
#donut chart for Region wise unit sold
plt.figure(figsize=(6,6))
Region_UnitsSold=df.groupby('Region')['Units Sold'].sum()
plt.pie(Region_UnitsSold,labels=Region_UnitsSold.index,autopct='%1.1f%%')

#create a donut chart using pie chart
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.title('Region wise Unit sold')
plt.show()
```

```python
#first groupby units sold by year and month then create a bar chart for units sold by year and month
UnitsSold_YearMonth=df.groupby(['Order Year','Order Month'])['Units Sold'].sum()
plt.figure(figsize=(10,4))
UnitsSold_YearMonth.plot(kind='bar')
plt.title('Units Sold by Year and Month')
plt.xlabel('Year and Month')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.show()

```

```python
#total cost by sales chennel
TotalCost_SalesChannel=df.groupby('Sales Channel')['Total Cost'].sum()
#pie chart for total cost by sales chennel
plt.figure(figsize=(6,6))
plt.pie(TotalCost_SalesChannel,labels=TotalCost_SalesChannel.index,autopct='%1.1f%%')
plt.title('Total Cost by Sales Channel')
plt.show()

```
